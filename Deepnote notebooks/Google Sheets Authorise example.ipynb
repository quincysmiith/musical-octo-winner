# Example of how to connect to Google Sheets 

Notebook running on Deepnote.

Contents of json service key were set as environment variables using this guide: [https://docs.deepnote.com/integrations/google-bigquery]

```
import json
import os
from google.oauth2 import service_account
import gspread
```

Set the scope of the credentials to be used to apply to Google Drive and Sheets

```
scopes = [
    'https://www.googleapis.com/auth/spreadsheets',
    'https://www.googleapis.com/auth/drive'
    ]
```

Set credentials by loading the required data from the environment variables

```
credentials = service_account.Credentials.from_service_account_info(json.loads(os.environ['MARQUIN_PERSONAL_GOOGLE_SERVICE_ACCOUNT']), scopes = scopes)
```

```
gc = gspread.authorize(credentials)
```

Open and read a spreadsheet

```
sh = gc.open('Weight Logs')

worksheet = sh.worksheet("Sheet1")

list_of_lists = worksheet.get_all_values()
```
