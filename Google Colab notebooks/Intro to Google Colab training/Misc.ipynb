## upload and download files

````
# upload files

from google.colab import files
files.upload()

# download files

from google.colab import files
files.download('path/to/your/file')
````

## Common bash commands

Download dataset from the web with 
```
!wget <ENTER URL>
```

Install libraries with 
```
!pip install <LIBRARY>
```
or 
```
!apt-get install <LIBRARY>
```
Run an existing .py script with 
```
!python script.py
```

Clone a git repository with 
```
!git clone <REPOSITORY URL>
```

## Magic commands

````
%lsmagic
````

run a file

````
%run hello.py
````

see environment variables

````
%env
````

load an external file

````
%load <file name>
````

run javascript code

````
%%javascript

prompt("hello python")
````

write html

````
%%html

<h1> a header </h1>

<h2> a smaller header </h2>

<img src="https://assets.pokemon.com/assets/cms2/img/pokedex/detail/001.png"> 
````
